```python
# Imports
import pandas as pd
import numpy as np
import zipfile
```

```python
# Unzip and read the data
filenames = ['force_plate', 'joint_angles', 'joint_velos', 'landmarks']
dfs = {}

for filename in filenames:
    with zipfile.ZipFile(f'../data/data/full_sig/{filename}.zip', 'r') as z:
        with z.open(f'{filename}.csv') as f:
            dfs[filename] = pd.read_csv(f)

force_plate_df = dfs['force_plate']
joint_angles_df = dfs['joint_angles']
joint_velos_df = dfs['joint_velos']
landmarks_df = dfs['landmarks']
```

```python
# View force plate data
print("Force Plate Data Shape:", force_plate_df.shape)
force_plate_df.head()
```

```python
# View joint angle data
print("Joint Angles Data Shape:", joint_angles_df.shape)
joint_angles_df.head()
```

```python
# View joint velocity data
print("Joint Velocities Data Shape:", joint_velos_df.shape)
joint_velos_df.head()
```

```python
# View landmarks data
print("Landmarks Data Shape:", landmarks_df.shape)
landmarks_df.head()
```

```python
# Merge joint angles with joint velocities
merged_df = joint_angles_df.merge(joint_velos_df, on=['time', 'session_swing'], suffixes=('', '_dup'))
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_dup')]

# Merge with landmarks
merged_df = merged_df.merge(landmarks_df, on=['time', 'session_swing'], suffixes=('', '_dup'))
merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_dup')]

# Merge with force plate data last
# merged_df = merged_df.merge(force_plate_df, on=['time', 'session_swing'], suffixes=('', '_dup'))
# merged_df = merged_df.loc[:, ~merged_df.columns.str.endswith('_dup')]
```

```python
# View merged data
print("Merged Data Shape:", merged_df.shape)
merged_df.head()
```

```python
# Collect eligible swings
eligible_swings_df = pd.read_csv('../eligible_swings.csv')
eligible_swings_df.head()
```

```python
# Left join with merged data
merged_df = eligible_swings_df.merge(merged_df, on=['session_swing'], how='left')

# Drop all columns from eligible_swings_df except for 'session_swing'
merged_df = merged_df.loc[:, ~merged_df.columns.isin(eligible_swings_df.columns[1:])]
```

```python
# View the merged dataframe
print("Final Merged Data Shape:", merged_df.shape)
merged_df.head()
```

```python
# Find the episode length per session
def find_episode_length(session_swing):
    session_data = merged_df[merged_df['session_swing'] == session_swing]
    episode_length = session_data['time'].max() - session_data['time'].min()
    return episode_length

# Apply the function to each session_swing
episode_lengths = merged_df['session_swing'].unique()
episode_lengths_df = pd.DataFrame({
    'session_swing': episode_lengths,
    'episode_length': [find_episode_length(swing) for swing in episode_lengths]
})
```

```python
# View episode lengths
print("Episode Lengths Data Shape:", episode_lengths_df.shape)
episode_lengths_df.head()
```

```python
# Distribution of episode lengths
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(episode_lengths_df['episode_length'], bins=30, kde=True)
plt.title('Distribution of Episode Lengths')
plt.xlabel('Episode Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# Find number of states per swing session
def find_num_states(session_swing):
    session_data = merged_df[merged_df['session_swing'] == session_swing]
    num_states = session_data['time'].nunique()
    return num_states

# Apply the function to each session_swing
num_states = merged_df['session_swing'].unique()
num_states_df = pd.DataFrame({
    'session_swing': num_states,
    'num_states': [find_num_states(swing) for swing in num_states]
})
```

```python
# View number of states
print("Number of States Data Shape:", num_states_df.shape)
num_states_df.head()
```

```python
# Distribution of states
sns.histplot(num_states_df['num_states'], bins=30, kde=True)
plt.title('Distribution of Number of States')
plt.xlabel('Number of States')
plt.ylabel('Frequency')
plt.show()
```

```python
# Save the merged dataframe to a CSV file
merged_df.to_csv('../rh_swing_data.csv', index=False)
```
